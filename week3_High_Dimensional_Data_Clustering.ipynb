Start from a public data set about people with hobbies.
### Task 1: Data Visualization, Distance Metrics

```python
from yellowbrick.datasets import load_hobbies
```

### Load the Corpus

```python
corpus = load_hobbies()
```

```python
print(corpus) #contains categories, files, data
```

```python
print(corpus.data)
```

### Vectorize the Documents

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
docs = vectorizer.fit_transform(corpus.data)
labels = corpus.target
label_vals=corpus.categories
```

```python
print('The shape of data=',docs.shape)
```

### t-SNE: Clustering Similar Documents with Squared Euclidean Distance and Euclidean Distance

### L2 Squared Euclidean Distance

```python
from yellowbrick.text import TSNEVisualizer

tsnse = TSNEVisualizer(size = (600, 400)) # L2 Squared Euclidean Distance
tsnse.fit(docs, labels)
tsnse.poof()
```

### Euclidean Distance

```python
tsnse = TSNEVisualizer(size = (600, 400), metrics = "euclidean") # Euclidean Distance
tsnse.fit(docs, labels)
tsnse.poof()
```

### Manhattan ("Taxicub" or "Cityblock") Distance

```python
tsne = TSNEVisualizer(size = (600, 400), metrics = "citycub")
tsne.fit(docs, labels)
tsne.poof()
```

### Other distance metrics include: Bray Curtis Dissimilarity, Canberra Distance, Cosine Distance, Hamming Distance

### Task 2: Data dimensionality reduction 
### Lets reduce data dimensions to 2D and visualize the same
### First, lets use PCA using TruncatedSVD to take care of sparsity in data

```python
from sklearn.decomposition import TruncatedSVD
import numpy as np
np.random.seed(19680801)
#Reducing the data to 2 dimensions to improve computation and visualization
clf = TruncatedSVD(2)
pca_docs=np.array(clf.fit_transform(docs))
labels=np.array(labels)
label_vals=np.array(label_vals)
print('Reduced data dimensions=', pca_docs.shape)
#print(labs)
```

```python
#Now create a function to visualize reduced dimension data

import matplotlib.pyplot as plt
import matplotlib.cm as cm

def visualize_data(x,y,cats,lbl):
  
  fig, ax = plt.subplots(figsize=(8, 6))
  colors = cm.rainbow(np.linspace(0, 1, len(cats)))
  
  for cat, co in zip(cats, colors):
      loc=np.where(y==cat)
      ax.scatter(x[loc,0], x[loc,1], color=co,label=cat)
  ax.set_xlabel('Component 1')
  ax.set_ylabel('Component 2')
  ax.set_title(lbl)
  ax.legend()
  ax.grid(False)

  plt.show()
```

```python
#Visualize 
visualize_data(pca_docs,labels,label_vals,'PCA Components')
```

# Next, Apply t-sne for feature reduction

```python
import sklearn.manifold
tsne_op = sklearn.manifold.TSNE(n_components=2, perplexity=2
                                )
docs_tsne = np.array(tsne_op.fit_transform(docs))
```

### Document on how to use perplexity https://distill.pub/2016/misread-tsne/

```python
#visualize the output
visualize_data(docs_tsne,labels,label_vals,'t-SNE Components')
```

### Task 3: Apply Yellowbrick library for Elbow method to identify optimal number of clusters

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,10))

visualizer.fit(pca_docs)        # Fit the data to the visualizer
```

```python
# 6 Optimal clusters from previous graph. Lets plot the 6 clusters
from sklearn.cluster import KMeans

clusters = KMeans(n_clusters = 6)
C_vals=np.array(clusters.fit_predict(pca_docs))
print(C_vals.shape)
labs=np.unique(C_vals)
visualize_data(pca_docs,C_vals,labs,'k-means on PCA')
```

```python
model = KElbowVisualizer(KMeans(), k=(1,10))
model.fit(docs_tsne)
```

```python
# 6 Optimal clusters from previous graph. Lets plot the 6 clusters
from sklearn.cluster import KMeans

clusters = KMeans(n_clusters = 5)
C_vals=np.array(clusters.fit_predict(docs_tsne))
print(C_vals.shape)
labs=np.unique(C_vals)
visualize_data(docs_tsne,C_vals,labs,'k-means on PCA')
```
